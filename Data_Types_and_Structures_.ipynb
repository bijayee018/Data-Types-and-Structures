1. What are data structures, and why are they important?
:-
Data structures are specialized formats for organizing, storing, and processing data efficiently to enable faster access, retrieval, and manipulation.

This is important because Efficient data managment,optimized performance,code redability and maintainbility,handaling of complex data items.

2. Difference between mutable and immutable data types (with examples)

Mutable: Can be changed after creation.

Example: list1 = [1,2,3]; list1[0] = 99 → [99,2,3].

Immutable: Cannot be changed once created.

3. Lists vs Tuples in Python

List: Mutable, slower, uses [].

Tuple: Immutable, faster, uses ().

Example:

my_list = [1, 2, 3]   # can change
my_tuple = (1, 2, 3)  # cannot change

4. How dictionaries store data

Dictionaries store data as key-value pairs using hashing.

Example: student = {"name": "Alex", "age": 21}

5. Why use a set instead of a list?

Sets remove duplicates automatically.

Faster membership testing (in).

Example:

s = {1,2,2,3}   # {1,2,3}

6. What is a string in Python, and how is it different from a list?

String: Immutable sequence of characters.

List: Mutable sequence of elements (any type).

Example: "hello" vs ['h','e','l','l','o'].

7. How do tuples ensure data integrity?

Since tuples are immutable, data cannot be altered by mistake → good for constants.

8. What is a hash table, and how does it relate to dictionaries?

Hash table: A data structure that maps keys to values using a hash function.

Python dictionary is implemented as a hash table.

9. Can lists contain different data types?

 Yes.

mixed = [1, "hello", 3.5, True]

10. Why are strings immutable in Python?

For memory efficiency and security.

Allows hashing (strings can be dictionary keys).

11. Advantages of dictionaries over lists

Fast lookup by key (O(1) average).

Lists require searching element-by-element (O(n)).

12. Scenario where tuple is preferable over list

When storing fixed data that should not change.

Example: (latitude, longitude) for coordinates.

13. How do sets handle duplicate values?

Duplicates are automatically removed.

14. How does the “in” keyword work differently for lists and dictionaries?

List: Checks values.

2 in [1,2,3]   # True


Dictionary: Checks keys only.

"name" in {"name":"Alex"}   # True
"Alex" in {"name":"Alex"}   # False

15. Can you modify the elements of a tuple? Why or why not?

:- No, tuples are immutable.

t = (1,2,3)
t[0] = 9  # Error

16. What is a nested dictionary? Example use case

Dictionary inside another dictionary.

students = {
    "A101": {"name": "Alex", "marks": 85},
    "A102": {"name": "Sam", "marks": 90}
}

17. Time complexity of accessing elements in a dictionary

Average: O(1) (constant time, due to hashing).

Worst-case: O(n) (if many collisions).

18. When are lists preferred over dictionaries?

When order matters or data doesn’t need a key.

Example: maintaining a sequence of steps.

19. Why are dictionaries considered unordered?

Before Python 3.7 → no guaranteed order.

From Python 3.7+ → maintains insertion order, but logically retrieval is key-based, not position-based.

20. Difference between list and dictionary in terms of data retrieval

List: Retrieved by index (mylist[0]).

Dictionary: Retrieved by key (mydict["name"]).
